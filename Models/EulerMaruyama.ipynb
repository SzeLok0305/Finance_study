```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def black_scholes_simulation(S0, r, q, sigma, T, N, num_sims):
    #Model: dS = (r-q) * S * dt + sigma * S * dW
    # Remember to use ito calculus  !!!!!!!
    dt = T / N
    paths = np.zeros((num_sims, N+1))
    paths[:, 0] = S0
    
    for i in range(1, N+1):
        #The solution take the form: S_t = S_0 * exp( (r - q - 1/2 * sigma^2) * t + sigma * Normal(0,sqrt(t)) )
        
        dW = np.random.normal(0, np.sqrt(dt), num_sims)
        paths[:, i] = paths[:, i-1] * np.exp((r - q - 0.5 * sigma**2) * dt + sigma * dW)
    
    return paths

def jump_diffusion_simulation(S0, r, q, sigma, lambda_j, mu_j, sigma_j, T, N, num_sims):
    #Model: dS = (r-q) * S * dt + sigma * S * dW + S * dN(lambda_j) * dJ(mu_j,sigma_j)
    # Remember to use ito calculus  !!!!!!!
    dt = T / N
    paths = np.zeros((num_sims, N+1))
    paths[:, 0] = S0
    
    for i in range(1, N+1):
# The solution take the form: S_t = S_0 * exp( (r - q - 1/2 * sigma^2) * t + sigma * Normal(0,sqrt(t)) + Poisson(lambda_j * dt) * Normal(mu_j,sigma_j) )
        
        # Diffusion component
        dW = np.random.normal(0, np.sqrt(dt), num_sims)
        diffusion = (r - q - 0.5 * sigma**2) * dt + sigma * dW
        
        # Jump component
        dN = np.random.poisson(lambda_j * dt, num_sims)
        J = np.random.normal(mu_j, sigma_j, num_sims)
        jump = J * dN
        
        #Combine
        paths[:, i] = paths[:, i-1] * np.exp(diffusion + jump)
        
    return paths
```

```python
S0 = 100 # Initial stock price
# The following parameters are annualized
r = 0.05 # Risk-free rate
q = 0.01 # Dividend
sigma = 0.15 # Volatility(%)
T = 1 # Time to maturity (1 year)
N = int(T*252) # Number of time steps (e.g. trading days in a year)
num_sims = 5 # Number of simulation paths

# Run simulation for Black-Scholes
sims_BS = black_scholes_simulation(S0, r, q, sigma, T, N, num_sims)
```

```python
#Preper for jump diffusion
lambda_j = int(N*0.01) # Jump intensity (average # of jump per year)
mu_j = -0.05  # mean jump size (%)
sigma_j = 0.20 # Volatility of jump size (%)

#Run simulation for jump diffusion
sims_JD = jump_diffusion_simulation(S0, r, q, sigma, lambda_j, mu_j, sigma_j, T, N, num_sims)
```

```python
def plot_result(sims,model_name):
    plt.figure(figsize=(10, 6))
    for i in range(len(sims)):
        plt.plot(np.linspace(0, T, N+1), sims[i])
    plt.xlabel('Time (years)')
    plt.ylabel('Stock Price')
    plt.title( f"{model_name} Model" )
    plt.grid(True)
    
plot_result(sims_BS,"Black-Scholes")
plot_result(sims_JD,"Jump diffusion")
```

```python
#Analys the averaged behaviour
def Plot_average(num_sims,sim_conditions):
    sim_conditions = S0, r, q, sigma, lambda_j, mu_j, sigma_j, T, N
    BS = black_scholes_simulation(S0, r, q, sigma, T, N, num_sims)
    JD = jump_diffusion_simulation(S0, r, q, sigma, lambda_j, mu_j, sigma_j, T, N, num_sims)

    BS_avg = np.mean(BS, axis=0)
    JD_avg = np.mean(JD, axis=0)

    time_step = np.linspace(0, T, N+1)
    plt.figure(figsize=(10, 6))
    plt.plot(time_step, BS_avg,label = "Black-Scholes averaged")
    plt.plot(time_step, JD_avg,label = "Jump diffusion averaged")
    plt.legend()
    plt.xlabel("Stock price")
    plt.show()

    #Plot final price distibution
    BS_distribution = np.log(BS[:,-1]/BS[:,0])
    JD_distribution = np.log(JD[:,-1]/JD[:,0])
    plt.figure(figsize=(10, 6))
    plt.title("log return spread")
    plt.hist(BS_distribution, label = "Black-Scholes",bins="auto",alpha=0.3,density=True)
    plt.hist(JD_distribution, label = "Jump diffusion",bins="auto",alpha=0.3,density=True)
    plt.legend()
    plt.xlabel("Stock price")
    plt.show()

    return [BS_avg,JD_avg], [BS_distribution,JD_distribution]

initial_para = [S0, r, q, sigma, lambda_j, mu_j, sigma_j, T, N]
Avg, Dis = Plot_average(2**14,initial_para)
```
