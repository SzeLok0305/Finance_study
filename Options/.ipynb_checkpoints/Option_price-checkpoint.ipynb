```python
import numpy as np
import pandas as pd
import yfinance as yf
import pandas_datareader as pdr
import seaborn as sns
from matplotlib import pyplot as plt
from scipy.stats import norm
```

Define the black scholes pricing model
Spot price: S
Strike: K
Maturaty: T
risk-free rate: r
volatility: sigma

```python
def black_scholes(S, K, T, r, sigma, option_type='call'):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    
    if option_type == 'call':
        option_price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    elif option_type == 'put':
        option_price = K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
    else:
        raise ValueError("Invalid option type. Use 'call' or 'put'.")
    
    return option_price
```

```python
#Get real world data (Stock price and risk-free rate)
Start_date = "2020-01-01"
df = yf.download("TSLA",start = Start_date)
df_bonds = pdr.DataReader("DGS3MO", 'fred', Start_date).dropna()
df_bonds["DGS3MO"] = df_bonds["DGS3MO"]/100 #Correcting the numerical value
```

Define a window (in # of days) to calculate the volatility.
Since we are using the 3-month US bond as the rfr, we set 90-days as default.

```python
df['Returns'] = np.log(df['Close'] / df['Close'].shift(1))
window = 90
df["vol"] = df["Returns"].rolling(window=90).std()
df["vol"] =df["vol"]*np.sqrt(252/window) #Annualize to 252 trading days = 1 year
df = pd.concat([df, df_bonds], axis=1)
df = df.dropna()
```

Create a overlapping plot to display the price of the option with different strike and Maturaty.
We will use the Open as the spot price and the "vol" as volatility

```python
def Plot_price(dataframe):
    Spot = np.array(dataframe["Open"])
    volatility = np.array(dataframe["vol"])
    risk_free_rate = np.array(dataframe["DGS3MO"])
    print(len(Spot),len(volatility),len(risk_free_rate))
    Call = black_scholes(Spot, 100, 1, risk_free_rate, volatility, option_type='call')
    return Call
```

```python
Price_shift = [-0.2,0,+0.2]
for O in ["call","put"]:
    for X in Price_shift:
        Name = O + str(X) if X != 0 else O + "ATM"
        df[Name] = black_scholes(df["Open"], df["Open"]*(1+X), 1, df["DGS3MO"], df["vol"], option_type=O)
```

```python
df["call-0.2"].plot(label="Call -20% spot price")
df["put0.2"].plot(label="Put +20% spot price")
plt.legend()
plt.ylabel("Option price")
plt.show()

df["call0.2"].plot(label="Call +20% spot price")
df["put-0.2"].plot(label="Put -20% spot price")
plt.legend()
plt.ylabel("Option price")
plt.show()

df["callATM"].plot(label="Call ATM")
df["putATM"].plot(label="Put ATM")
plt.legend()
plt.ylabel("Option price")
plt.show()
```
