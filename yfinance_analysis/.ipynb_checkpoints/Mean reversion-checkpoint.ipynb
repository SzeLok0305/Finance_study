```python
import yfinance as yf
import pandas as pd
import numpy as np
import ta
import matplotlib.pyplot as plt
```

```python

df =yf.download("NVDA", start = "2024-01-01")
```

```python
df["ma_20"] = df.Close.rolling(20).mean()
df["vol"] = df.Close.rolling(20).std()
df["upper_bb"] = df.ma_20 + (2*df.vol)
df["lower_bb"] = df.ma_20 - (2*df.vol)
df["Shifted_Close"] = df.Close.shift() # For Stop loss
```

```python
df[["Close","ma_20","upper_bb","lower_bb"]].plot()
```

```python
df["rsi"] = ta.momentum.rsi(df.Close,window = 6) #relative strength index
conditions = [(df.rsi < 30) & (df.Close < df.lower_bb), (df.rsi > 70) & (df.Close > df.upper_bb)]
choies = ["buy", "sell"]
df["Signal"] = np.select(conditions, choies)
```

```python
df.dropna(inplace = True)
#Shift 1 roll for "Signal" since you can only buy on the next "Open"
df.Signal = df.Signal.shift()
```

```python
position = False
buydates, selldates = [], []
buyprice, sellprice = [], []
stop_loss_fraction = 0.9

for index, row in df.iterrows():
    if not position and row["Signal"] == "buy":
        buydates.append(index)
        buyprice.append(row.Open)
        position = True

    if position:
        if row["Signal"] == "sell" or row.Shifted_Close < stop_loss_fraction*buyprice[-1]:
            selldates.append(index)
            sellprice.append(row.Open)
            position = False
```

```python
plt.figure(figsize=(10,5))
plt.plot(df.Close)
plt.scatter(df.loc[buydates].index,df.loc[buydates].Close, marker = "^", c= "g")
plt.scatter(df.loc[selldates].index,df.loc[selldates].Close, marker = "v", c= "r")
```

```python
Return = (pd.Series([(sell-buy)/buy for sell, buy in zip(sellprice,buyprice)]) + 1).prod() -1
print("% return = ", Return)
```
